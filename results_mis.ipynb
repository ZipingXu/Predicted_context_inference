```python
# test the effects of -l
mis_list = [True, False]
env_list = ['failure1', 'failure2', 'random_0.1_0', 'polynomial', 'neural_network']

# Create a 3x3 grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Counter for subplot position
plot_idx = 0

for env in env_list:
    for mis in mis_list:
        name = f'runs/mis/history_dict_{env}_mis_{mis}.pkl'
        print(name)
        if not os.path.exists(name):
            continue
        with open(name, 'rb') as f:
            history_dict, args = pickle.load(f)
        # Get the current subplot
        ax = axes[plot_idx]
        
        # Pass the axis to plot_coverage
        plot_coverage_w_ax(history_dict, args, ax=ax)
        
        # Add title to each subplot
        ax.set_title(f'Environment: {env}, mis: {mis}')
        
        # Increment plot index
        plot_idx += 1

# Adjust layout
plt.tight_layout()
plt.show()


```
